```python
import matplotlib. pyplot as plt
import numpy as np
```

```python
def square_and_plot (x_values):
    """
    Takes x values , square them to get y values and plots x vs y.

    parameters:
    x_values (list or array) : Input x values 
    tuple:(x_values,y_values) where y_values=x_values**2
    """
    x_array= np.array(x_values) # Convert input to NumPy array

    y_values=x_array**2  # Compute y = x²

    # Create the plot
    plt.figure(figsize=(6,8))
    plt. plot (x_array, y_values, 'b-', linewidth=2 , label='y=x²')
    plt.scatter(x_array, y_values, color='red', s=30, zorder=3)

    # Labels and title
    plt.xlabel("x_values")
    plt.ylabel("y_values(x²)")
    plt.title("plot of y = x²")

    # Labels and title
    plt.legend()
    plt.show() 
    
    return x_array, y_values
    

```

```python
x_smooth = np.linspace(-5, 5, 100) # Generate 100 equally spaced points between -5 and 5 for smooth plotting
x_smooth_result, y_smooth_result = square_and_plot(x_smooth ) # Call the function and save the output x values and their squared results
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rms_velocity(interval_velocities, time_intervals):
    """
    RMS Velocity Calculation (CMP Analysis)

    Mathematical Formula:
    Vrms = sqrt( Σ(Vint^2 * Δt) / Σ(Δt) )

    interval_velocities : interval velocities (m/s)
    time_intervals      : time intervals (s)

    return:
    Vrms : RMS velocity (m/s)
    """
    Vrms = np.sqrt(
        np.sum(interval_velocities**2 * time_intervals) /
        np.sum(time_intervals)
    )
    return Vrms

# Given CMP Data (Example)

# Interval velocities picked from CMP (m/s)
Vint = np.array([1800, 2200, 2600, 3000])

# Corresponding time intervals (s)
dt = np.array([0.4, 0.4, 0.4, 0.4])

# Two-way travel time (cumulative)
twt = np.cumsum(dt)  
# RMS Velocity Calculation
Vrms_value = rms_velocity(Vint, dt)
print("RMS Velocity =", Vrms_value, "m/s")

# RMS velocity picks for plotting (example CMP picks)
Vrms_picks = np.array([1800, 2100, 2400, 2700])

# CMP Velocity Plot
plt.figure(figsize=(6, 8))
plt.plot(Vrms_picks, twt, marker='o', linewidth=3, label="Vrms Picks")
plt.gca().invert_yaxis()
plt.xlabel("RMS Velocity (m/s)")
plt.ylabel("Two-Way Time (s)")
plt.title("CMP Velocity Analysis (Vrms vs TWT)")
plt.grid(True)
plt.legend()
plt.show()
```

**Language Structure and Meaning**

This setion will briefly cover the syntax and semantics of Python programming language which makes it easier to interpret, implement and dynamically code for multiple programming paradigms. For an in-depth details behind any of these concepts, the readers can refer to Rossum (2012), who is the author of Python programming language. Python being a simple and consistent in design is based on the philosophy of "There should be oneand preferably only oneobvious way to do it", as mentioned in "The Zen of Python" - https://www.python.org/dev/peps/pep-0020/. 

The structure or form of a language is what constitutes the syntax for the programming language and is usually comprised of three levels - words, phrases and context which defines the rules in which a program is written. No information about the meaning or execution of a command is provided with the syntax. On the other hand, semantics of a language is defined as the meaning behind the code and what is it supposed to provide as an outcome. Some of the rules and meaning behind the language are listed below:

* Python is a case sensitive language. Example: test_variable and Test_Variable are treated as two different objects.
* Unlike some other programming languages, Python does not require a command terminator. Example, use of semicolon ; after every statement is not required. However, if two different statements need to be executed in the same line of code, they can be separated by a semicolon. Example: 

```python
test_var = 10; print (test_var) # Assign the value 10 to the variable test_var
```

```python
x1 = 10*2
x2 = 10 * 2
x3 = 10   *    2

print(x1)
print(x2)
print(x3)
```

**Is everything an object in Python?**

Python being an object-oriented programming language treats everything as an object. With a  more loose definition of an 'object' in Python, as compared to other programming languages, some objects neither have attributes nor methods. Neither are all the objects subclassable. The creator of the language, Rossum (2009) - http://python-history.blogspot.com/2009/02/first-class-everything.html wanted 'all objects that could be named in the language (e.g., integers, strings, functions, classes, modules, methods, etc.) to have an equal status', and hence classified them as 'first class everything'. In other words, every object can be assigned to variables, included as a member of a list, used as key-value pair in a dictionary, or even possible to be passed as an argument. 

One of the advantages of Python being a dynamically typed and interpreted language, is the simplistic nature in which the variables are defined. In this case data types are checked as well as the source code is translated during the execution of the program. 

In Python, each variable name is bound only to an object. Usually assignment operators are used to carry out this operation. When a value is assigned to a variable, in many programming languages it is thought to be stored in the form of a container or a memory bucket (VanDerPlas, 2016). In Python this is better represented as a pointer instead of being a container. This is one of the reasons where there is no declaration of a 'datatype', which is very common in other languages. **Lists**
Enclosed within square brackets [], lists are an ordered and mutable array of elements, separated by a comma, which can include different data types like int, float, string etc. It is possible to have duplicate elements in a list. Nested list is a 'list of lists' and is often used for two-dimensional matrix. Lists and its elements could be accessed in different ways. The indexing starts from 0 which is the first element of the list. Negative indexes could also be used to access the elements of a list where the last element has an index of -1, the second last is -2 and so on. There are multiple methods available to access or modify elements of a list. Some of the examples include append(), pop(), count(), sort () and many more. Since the lists are mutable, it is possible to modify/change the elements of a list as well as combine multiple lists. 

```python
# lists 
list1 = [1,2,3] # simple list with integer values
list2 = [1,None,'Python',3.14] # list with mixed data types
list3 = [ [1,2,3], [10,20,30], ['Example', 'Nested', 'List']] # nested list
list4 = [1,2,3,3,2,2,2,1,5,8,0] # list with repeated elements
```

```python
print(list1[0]) # output is 1
print(list2[-2]) # output is Python
print(list3[2][1]) # output is 'Nested'
print(list4.count(2)) # output is 4
print(list3[2][1:3]) # slicing to access range of elements, output is ['Nested','List']
print(list1 + list2) # combining elements of two lists, output is [1, 2, 3, 1, None, 'Python', 3.14]
```

**Tuples** <br>
Enclosed within curved brackets (), tuples are very similar to a list, with a primary difference being their immutable nature (unchangeable). Tuples usually contain a heterogeneous sequence of elements accessed via unpacking or indexing. The elements are ordered and also separated by a comma, just like a list. Elements are accessed in the same way as a list and the indexing starts with 0 as well. Similar to lists, negative indexes can be used to access the elements of a tuple where the last element has an index of -1, the second last is -2 and so on. Some of the methods available for a tuple include count() and index(). Advantages of using a tuple instead of a list include fast execution compared to an equivalent list as well as situations where we do not want the data to be modified. 


```python
# tuples
tuple1 = (1,2,3) # simple tuple with integer values
tuple2 = (1,None,'Python',3.14) # tuple with mixed data types
tuple3 = ([1,2,3], [10,20,30], ['Example', 'Nested', 'Tuple']) # nested tuple
tuple4 = (1,2,3,3,2,2,2,1,5,8,0) # list with repeated elements
```

```python
print(tuple1[0]) # output is 1
print(tuple2[-2]) # output is Python
print(tuple3[2][1]) # output is 'Nested'
print(tuple4.count(2)) # output is 4
print(tuple3[2][1:3]) # slicing to access range of elements, output is ['Nested','List']
print(tuple1 + tuple2) # combining elements of two lists, output is [1, 2, 3, 1, None, 'Python', 3.14]
```

**Dictionary** <br>
Enclosed within curly brackets {}, dictionary is another composite datatype which is also sometimes known as associative array, and consists of a collection of key-value pairs. It has certain similarities with a list since both datatypes are mutable, dynamic (increase or decrease size) and can be nested (dictionary within a dictionary). The primary difference lies in the way elements are accessed. Dictionary elements are accessed via keys instead of the index position. A dictionary can also be created using a built-in function, dict(), as shown in the example below. One of the restrictions for a dictionary is that a given 'key' can only appear once. Some of the commonly used methods for a dictionary include clear(), items(), keys(), values() and pop(). 


```python
# dictionary

dict1 = {
    'Texas': 'Permian Basin',
    'Colorado': ' DJ Basin',
    'North Dakota': 'Williston Basin'
}

# nested dictionary
dict2 = dict(
    USA = {'Exxon', 'Chevron', 'Conoco'},
    UK = 'BP',
    Revenue = [279.3, 158.9, 38.73, 303.7])
```

```python
print(dict1)
print(dict2)
```

```python
print(dict1['Colorado']) # access an element using key, output is 'DJ Basin'
dict1['Texas'] = 'Eagle Ford' # assign a new value based on a key
print(dict1)
print(dict2['Revenue'][-1]) # Revenue of BP in billion, USD
print(list(dict2.items())[2][1]) # returns the Revenues. First item is key and second item is value
print(dict2.keys()) # output is a list of all keys
print(dict2.values()) # output is the value of all keys
```

**2.3 Control Flow Tools and Iterators**

Control flow is when certain block of code is required to perform an operation in a repeated manner. Some of the simple conditional statements include if, elif and else which uses boolean condition to execute a set of command.  Implementation of these concepts is illustrated using an example where we assume a simple case with only three types of rock permeability category. The use of indentation and semi-colon (:) is important to note here which is a part of Python syntax. 

```python
# example of if-elif-else
perm = 0.001 # input in millidarcy

if (perm > 0.0001):
    print('This is a millidarcy case, seems to be a sandstone or limestone rock!')
elif(perm <= 0.0001 and perm>= 1E-6):
    print('This is a microdarcy case, seems to be tight sand rock!')
else:
    print('This is a nanodarcy case, seems to be shale rock!')
```

On the other hand, there are two primary categories for loop controls statements which are either based on loop counts or controlled by an event. Examples of the loop controls include for, while, break, continue or pass. A 'for' statement belongs to the first category where loop is executed for every item in a list or a loop count. A 'while' statement is executed while a particular condition is true. There are cases when the code needs to be forced out of the loop and 'break' is used in that situation. 'Continue' is used when a remainder of a current loop needs to be skipped to another iteration. 

```python
for i in range(5):
    print('Entered the loop for count', i)
print('Loop Execution is complete!')
```

```python
while (perm<1E-6):
    print('Very Low Permeability typical for unconventional wells!')
else:
    print('Not an unconventional well!')
```

**2.4 Basic Object Oriented Programming**

While working on a project, often situations are encountered where a block of code is required to be executed in a repeated manner with different arguments. The code is organized in a much more reusable pieces which are known as 'functions'. In Python, they could be created by using def or lambda statement. An example of a case where a function can be used to convert units is shown below. Every time we need to convert the units again, we can simply call the function and pass the new argument. 

```python
# function to convert units for mud weight
def unit_conversion(a):
    '''
    This function is used to convert units of mud weight.
    
    Input: Mud Weight in kg/m3
    Output: Mud Weight in ppg or psi/ft or specific gravity
    
    '''
    print('The entered mud weight in kg/m3 is:', a) # Print the input mud weight value in kg/m3
    mw_ppg = a * 0.0083454 # Convert mud weight from kg/m3 to pounds per gallon (ppg)
    mw_psift = a * 0.000434 # Convert mud weight from kg/m3 to pressure gradient (psi/ft)
    sp_gravity = a * 0.001 # Convert mud weight from kg/m3 to specific gravity
    print('Mud Weight in ppg:', mw_ppg) # Print the mud weight value in ppg
    print('Mud Weight in psi/ft:', mw_psift)
    print('Specific Gravity:', sp_gravity)
```

```python
# enter the input for mud weight in kg/m3
mud_weight = 1250
unit_conversion(mud_weight)
```

When we are using only functions as blocks of statements within a code, it is called procedure-oriented way of programming. However there is another way of writing more complicated programs commonly known as object-oriented programming (OOP). Data and functionality (i.e. properties and behavior) is wrapped inside a class and object which are the two main aspects of OOP. Class is used to create a user-defined data structures with information related to a particular characteristic. It can be thought of as a blueprint which can be filled out with different instances. Different objects are created within a class which contain attributes. An example of a simple class is shown below. For a more detailed description of classes the readers are recommended to refer to https://docs.python.org/3/tutorial/classes.html. 

```python
class well():
    # class attribute
    well_type = 'oil well'
    
    # initializer
    def __init__(self, api, basin, oil_prod, gas_prod):
        self.api = api
        self.basin = basin
        self.oil_prod = oil_prod # BBL
        self.gas_prod = gas_prod # MCF
        
    # instance method
    def boe_calc(self):
        boe = self.oil_prod + self.gas_prod/5.8
        return boe        
```

```python
# instantiating the class
well_1 = well(12345, 'Williston', 1000, 5000)

# calling a method of the class
print(well_1.boe_calc()) # boe production of well_1

well_2 = well(78910, 'Permian', 2000, 6000)
print(well_2.boe_calc()) # boe production of well_1
```

```python
# access the instance attributes
print('API for first well is:', well_1.api, 'and its in', well_1.basin, 'Basin')
if well_1.well_type == 'oil well':
    print("{0} belongs to {1} category".format(well_1.api, well_1.well_type))
```

To merge the datasets from different sources, it is possible to combine them by multiple ways. A brief description is provided below. 

syntax: pd.merge(left, right, how, on)

Here, left and right indicate the first and second dataframe which are required to be merged

1. how = 'outer', This allows to take the union of datasets and append them together.
2. how = 'inner', This allows to take the intersection of datasets and keep only the common values
3. how = 'left', All indexes of left dataframe are included. If there are indexes which are not present in the 'left' dataframe, they will not be included in the merged dataset
4. how = 'right', All indexes of right dataframe are included. If there are indexes which are not present in the 'right' dataframe, they will not be included in the merged dataset

The keyword 'on' decides if the dataframes are merged on a particular ID which is a unique identifier. Eg. API Number in oil and gas wells.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def hydrostatic_pore_pressure(pw, g, h):
    """
    Calculates hydrostatic pore pressure.

    pw : mud weight in pounds per gallon (ppg)
    g  : gravity factor (unitless)
    h  : true vertical depth in feet (ft)

    Returns:
    Hydrostatic pore pressure in psi
    """

    conversion_constant = 0.052
    HHP = pw * g * h * conversion_constant
    return HHP


# Input values
mud_weight = 10   # ppg
gravity = 1       # unitless
depth = 5000      # ft

# Function call
pressure = hydrostatic_pore_pressure(mud_weight, gravity, depth)

print(f"Hydrostatic Pore Pressure = {pressure:.2f} psi")
```

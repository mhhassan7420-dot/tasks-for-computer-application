```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calculation_vertical_stress(depths, bulk_density, factor=0.052):
    """
    Calculate Vertical Stress (Sv)

    depths        : array of depths in feet (ft)
    bulk_density  : bulk density in ppg
    factor        : unit conversion factor (default = 0.052)

    returns       : vertical stress in psi
    """
    Sv = factor * bulk_density * depths
    return Sv


# Step 1: Depth range (500â€“10,000 ft)
depths = np.arange(500, 10000, 500)

# Step 2: Constants
bulk_density = 21  # ppg

# Step 3: Calculate Vertical Stress
Sv = calculation_vertical_stress(depths, bulk_density)

# Step 4: Plotting
plt.figure(figsize=(8, 6))
plt.plot(depths, Sv, label="Vertical Stress (Sv)", linewidth=2)

plt.xlabel("Depth (ft)")
plt.ylabel("Vertical Stress (psi)")
plt.title("Vertical Stress vs Depth")
plt.grid(True)
plt.legend()
plt.show()
```
